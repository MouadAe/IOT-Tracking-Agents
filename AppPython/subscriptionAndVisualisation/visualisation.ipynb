```python
conn=pymysql.connect(host='localhost',port=int(3306),user='root',passwd='',db='agentstracking')
df1=pd.read_sql_query("SELECT * FROM gps_tracking ",conn)
df2=pd.read_sql_query("SELECT * FROM agent ",conn)
tracking_data = pd.DataFrame(df1)
tracking_data
```

```python
def getClientRow():
    indexOFClient=0
    for index, row in tracking_data.iterrows():
        if(row["id_agent"].rfind("GPS")!= -1):
            indexOFClient=index
    return tracking_data.loc[indexOFClient]

# print(getClientRow())
def getTheMinDistance():
    clientRow = getClientRow()
    isInisialised=False
    minKm = 0.0000 
    index_closed_taxi = -1
    for index, row in tracking_data.iterrows(): 
        if(row["id_agent"].rfind("GPS") == -1 and agent_data.loc[int(row['id_agent'])-1]['type']=='TAXI'
           and agent_data.loc[int(row['id_agent'])-1]['isFree']==1 ):
            loc1=(clientRow['latitude'],clientRow['longitude'])
            loc2=(row['latitude'],row['longitude'])
            if(isInisialised ==False):
                minKm = hs.haversine(loc1,loc2)
                isInisialised=True
            if(minKm >= hs.haversine(loc1,loc2)):
                minKm= hs.haversine(loc1,loc2)
                index_closed_taxi = index+1
#             print(row,agent_data.loc[int(row['id_agent'])-1])
            print("------------------------",hs.haversine(loc1,loc2),"--------------------------")
    return index_closed_taxi,minKm        

    
getTheMinDistance()  


# tracking_data.loc[19]['latitude']
#     folium.CircleMarker([row['latitude'], row['longitude']],
#                         radius=2,
#                         popup=row['id'],
#                         fill_color="#3db7e4", # divvy color
#                        ).add_to(map)
# taxi(lat,log)  client(lat,lon)=> ? km ,id_agentTaxi
```

```python
clientRow = getClientRow()
m = folium.Map(location=[clientRow['latitude'],clientRow['longitude']], zoom_start=15)
tooltip = "Click me!"
def mapTracer():
    getIndexClosedTaxi,getMinDistance=getTheMinDistance()
    print( getIndexClosedTaxi,getMinDistance)
    clientRow = getClientRow()
    # tacer le client
    folium.Marker([clientRow['latitude'], clientRow['longitude']], popup="your position ",
                              tooltip=tooltip,
                              icon=folium.Icon(color='lightgray', icon='crosshairs', prefix='fa')).add_to(m)
    for index, row in tracking_data.iterrows():
        if(row["id_agent"].rfind("GPS") == -1):
            agentRow = agent_data.loc[int(row['id_agent'])-1]
            if(agentRow['id']==getIndexClosedTaxi):
                #tracer closed taxi ----------------
                print("---Closed taxi----------------")
                folium.PolyLine([[clientRow['latitude'],clientRow['longitude']],
                [row['latitude'], row['longitude']]
                ]).add_to(m)
                folium.Marker([row['latitude'], row['longitude']], popup="Closed TAXI : "+str(round(getMinDistance,3))+" Km",
                              tooltip=tooltip,
                              icon=folium.Icon(color="green", icon='taxi', prefix='fa')).add_to(m)
                print(row,agentRow)
                print("-------------------------------")
            else:
                #tracer soit taxi(hold, free) or client 
                if(agentRow['type']=='TAXI' and agentRow['isFree']==1):
                    print("---Free taxi----------------")
                    folium.Marker([row['latitude'], row['longitude']], popup="Free TAXI",
                              tooltip=tooltip,
                              icon=folium.Icon(color="orange", icon='taxi', prefix='fa')).add_to(m)
                    print(row,agentRow)  
                    print("-------------------------------")
                if(agentRow['type']=='TAXI' and agentRow['isFree']==0):
                    print("---Occuped taxi----------------")
                    folium.Marker([row['latitude'], row['longitude']], popup="Occuped TAXI",
                              tooltip=tooltip,
                              icon=folium.Icon(color="red", icon='taxi', prefix='fa')).add_to(m)
                    print(row,agentRow)  
                    print("-------------------------------")
                if(agentRow['type']=='CLIENT'):
                    print("---client ----------------")
                    folium.Marker([row['latitude'], row['longitude']], popup="Client ",
                              tooltip=tooltip,
                              icon=folium.Icon(color="blue", icon='user', prefix='fa')).add_to(m)
                    print(row,agentRow)  
                    print("-------------------------------")                 
        
mapTracer()
    
```
